```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score

# Set up the plotting style and inline display
%matplotlib inline
sns.set()
sns.set_theme(style="white")

# Load data from Excel files
df_train_adaboost = pd.read_excel('./predicted_adaboost_model.xlsx', sheet_name='training')
df_test_adaboost = pd.read_excel('./predicted_adaboost_model.xlsx', sheet_name='testing')

df_train_gboost = pd.read_excel('./predicted_gradient_boosting_model.xlsx', sheet_name='training')
df_test_gboost = pd.read_excel('./predicted_gradient_boosting_model.xlsx', sheet_name='testing')

df_train_xgboost = pd.read_excel('./predicted_xgboost_model.xlsx', sheet_name='training')
df_test_xgboost = pd.read_excel('./predicted_xgboost_model.xlsx', sheet_name='testing')

df_train_ann_y_train = pd.read_csv('./ann files/y_train.csv')
df_train_ann_y_pred = pd.read_csv('./ann files/pred_1.csv')
df_test_ann_y_test = pd.read_csv('./ann files/y_test.csv')
df_test_ann_y_pred = pd.read_csv('./ann files/pred_2.csv')

df_train_stack_b = pd.read_excel('./stack b/grid_search_stacking_train_test_data.xlsx', sheet_name='training')
df_test_stack_b = pd.read_excel('./stack b/grid_search_stacking_train_test_data.xlsx', sheet_name='testing')

# Calculate R scores
r2_training_adaboost = r2_score(df_train_adaboost['y_train'], df_train_adaboost['y_train_pred'])
r2_test_adaboost = r2_score(df_test_adaboost['y_test'], df_test_adaboost['y_test_pred'])

r2_training_gboost = r2_score(df_train_gboost['y_train'], df_train_gboost['y_train_pred'])
r2_test_gboost = r2_score(df_test_gboost['y_test'], df_test_gboost['y_test_pred'])

r2_training_xgboost = r2_score(df_train_xgboost['y_train'], df_train_xgboost['y_train_pred'])
r2_test_xgboost = r2_score(df_test_xgboost['y_test'], df_test_xgboost['y_test_pred'])

r2_training_ann = r2_score(df_train_ann_y_train, df_train_ann_y_pred)
r2_test_ann = r2_score(df_test_ann_y_test, df_test_ann_y_pred)

r2_training_stack_b = r2_score(df_train_stack_b['y_train'], df_train_stack_b['y_train_pred'])
r2_test_stack_b = r2_score(df_test_stack_b['y_test'], df_test_stack_b['y_test_pred'])


# Set up font sizes for plots
SMALL_SIZE = 11
MEDIUM_SIZE = 12
BIGGER_SIZE = 14

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Times New Roman'
plt.rcParams['mathtext.it'] = 'Times New Roman'
plt.rcParams['mathtext.bf'] = 'Times New Roman'
plt.rcParams["font.family"] = 'Times New Roman'

# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(8, 9))

# Plot configurations
plot_configs = [
    (axes[0, 0], df_train_adaboost, df_test_adaboost, '(a) AdaBoost (AB)', r2_training_adaboost, r2_test_adaboost),
    (axes[0, 1], df_train_gboost, df_test_gboost, '(b) Gradient Boosting (GB)', r2_training_gboost, r2_test_gboost),
    (axes[1, 0], df_train_xgboost, df_test_xgboost, '(c) XGBoost (XGB)', r2_training_xgboost, r2_test_xgboost),
    (axes[1, 1], df_train_ann_y_pred, df_test_ann_y_pred, '(d) ANN', r2_training_ann, r2_test_ann),
    (axes[2, 0], df_train_stack_b, df_test_stack_b, '(e) Stack_B', r2_training_stack_b, r2_test_stack_b)
]

# Plotting residuals with different markers and colors for train and test data
for ax, train_data, test_data, title, r2_train, r2_test in plot_configs:
    if title == '(d) ANN':
        ax.scatter(train_data.to_numpy(), train_data.to_numpy() - df_train_ann_y_train.to_numpy(), 
                   c='red', marker="o", edgecolor='white', label='Train data')
        ax.scatter(test_data.to_numpy(), test_data.to_numpy() - df_test_ann_y_test.to_numpy(), 
                   c='limegreen', marker='s', edgecolor='white', label='Test data')
    else:
        ax.scatter(train_data['y_train_pred'], train_data['y_train_pred'] - train_data['y_train'], 
                   c='red', marker="o", edgecolor='white', label='Train data')
        ax.scatter(test_data['y_test_pred'], test_data['y_test_pred'] - test_data['y_test'], 
                   c='limegreen', marker='s', edgecolor='white', label='Test data')

    ax.set_ylabel('Residuals')
    ax.set_xlabel('Predicted values')
    ax.hlines(y=0, color='black', xmin=np.min(train_data), xmax=np.max(train_data), lw=1.0, linestyles='--')
    ax.set_title(title)
    
    # unnecessary code
    if title == "(a) AdaBoost (AB)":
        ax.legend(loc='lower right')
    elif title == "(c) XGBoost (XGB)":
        ax.legend(loc='lower left')
    else:
        ax.legend(loc='upper right')

# Create a separate subplot for R values
ax_r2 = axes[2, 1]
ax_r2.axis('off')

# Adding R values as text with enhanced formatting
r2_text_1 = f"""
AdaBoost (AB):
R (train) = {r2_training_adaboost:.3f}    R (test) = {r2_test_adaboost:.3f}
Gradient Boosting (GB):
R (train) = {r2_training_gboost:.3f}    R (test) = {r2_test_gboost:.3f}
XGBoost (XGB):
R (train) = {r2_training_xgboost:.3f}    R (test) = {r2_test_xgboost:.3f}
ANN:
R (train) = {r2_training_ann:.3f}    R (test) = {r2_test_ann:.3f}
Stack_B:
R (train) = {r2_training_stack_b:.3f}    R (test) = {r2_test_stack_b:.3f}
"""

# r2_text_2 = f"""
# ANN:
# R (train) = {r2_training_ann:.3f}    R (test) = {r2_test_ann:.3f}

# Stacking B:
# R (train) = {r2_training_stack_b:.3f}    R (test) = {r2_test_stack_b:.3f}
# """

ax_r2.text(0.1, 0.5, r2_text_1, fontsize=MEDIUM_SIZE, verticalalignment='center', horizontalalignment='left', transform=ax_r2.transAxes)
# ax_r2.text(0.5, 0.5, r2_text_2, fontsize=MEDIUM_SIZE, horizontalalignment='left', transform=ax_r2.transAxes)

plt.tight_layout()

# Save the figure with DPI 400
plt.savefig('residual_plots_and_r2_values.png', dpi=1000)

# Display the plot
plt.show()
```
